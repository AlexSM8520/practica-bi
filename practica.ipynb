```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

File_Exercise = "train.csv"

tb = pd.read_csv(File_Exercise)

#print(tb.head)

#we start to implementar estadistica

print(tb.describe())
```

```python

Top_5_Customers = tb.groupby('Customer ID')["Sales"].sum().nlargest(5)
Best_Customer = tb[tb['Customer ID'].isin(Top_5_Customers.index)]
Top_5_names_customer = Best_Customer[['Customer Name', 'Sales']]
print(Top_5_names_customer)

plt.bar(Best_Customer['Customer Name'], Best_Customer['Sales'],color='red')
plt.ylabel('Sales')
plt.title('Top 5 Customers by Sales')
plt.xticks(rotation=90)
plt.show()

```

```python
Frecuencia_Product = tb['Product ID'].value_counts()
print(Frecuencia_Product.head())
```

```python
Top_10_statesSales = tb.groupby("City")["Sales"].sum().nlargest(10)
Best_StatesSales = tb[tb['City'].isin(Top_10_statesSales.index)]
New_Cites = Best_StatesSales[['City', 'Sales']]


plt.scatter(New_Cites['City'],New_Cites['Sales'], alpha=0.5)
plt.title('ventas por Cuidad')
plt.xlabel('Ciudades')
plt.ylabel('Ventas')
plt.xticks(rotation=90)
plt.show()
print(Top_10_statesSales)
```
